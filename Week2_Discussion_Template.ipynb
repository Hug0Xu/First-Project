```python
female_df = df[df['Gender'] == 0].copy()
male_df = df[df['Gender'] == 1].copy()
female_df
```

```python
frank1 = sum(female_df['Rank']==1)
frank2 = sum(female_df['Rank']==2)
frank3 = sum(female_df['Rank']==3)
mrank1 = sum(male_df['Rank']==1)
mrank2 = sum(male_df['Rank']==2)
mrank3 = sum(male_df['Rank']==3)

fig, axes = plt.subplots(nrows = 3, ncols = 2, figsize=(40, 25))
pie_chart_x = [frank1,frank2,frank3]
pie_chart_x_males = [mrank1,mrank2,mrank3]
pie_chart_x_lab = ["female rank 1", "female rank 2", "female rank 3"]
pie_chart_x_male_lab = ["male rank 1","male rank 2", "male rank 3"]

#female ranks
axes[0,0].pie(pie_chart_x,labels= pie_chart_x_lab,autopct='%1.1f%%')
axes[0,0].set_title('Distribution of Ranks of females')
axes[0,0].legend()

#male ranks
axes[0,1].pie(pie_chart_x_males,labels= pie_chart_x_male_lab,autopct='%1.1f%%')
axes[0,1].set_title('Distribution of Ranks of ales')
axes[0,1].legend()

# female experience
axes[1,0].hist(x=female_df['Exper'])
axes[1,0].set_title('Female Distribution of Experience')
axes[1,0].set_xlabel('Experience (Years since obtaining MD)')
axes[1,0].set_ylabel('Count')

# males experience
axes[1,1].hist(x=male_df['Exper'])
axes[1,1].set_title('Male Distribution of Experience')
axes[1,1].set_xlabel('Experience (Years since obtaining MD)')
axes[1,1].set_ylabel('Count')

# females salary
axes[2,0].hist(x=female_df['Sal95'])
axes[2,0].set_title('Female Distribution of Salary after 1994')
axes[2,0].set_xlabel('Experience (Years since obtaining MD)')
axes[2,0].set_ylabel('Count')

# males salary
axes[2,1].hist(x=male_df['Sal95'])
axes[2,1].set_title('Male Distribution of Salary after 1994')
axes[2,1].set_xlabel('Experience (Years since obtaining MD)')
axes[2,1].set_ylabel('Count')

plt.tight_layout()
plt.show()
```
